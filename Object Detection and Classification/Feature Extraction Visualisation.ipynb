```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
```

```python
img = cv2.imread('train/train_positive_A/00000000/item_00000002.pnm')
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
feature_matrix, hog_image = hog(img, orientations=9, pixels_per_cell=(8, 8),cells_per_block=(2, 2), visualize=True, multichannel=True)
```

```python
plt.imshow(hog_image)
```

```python
neg_img = cv2.imread('train/train_negative_A/00000000/item_00000000.pnm')
```

```python
plt.imshow(neg_img)
```

```python
neg_img.shape
```

```python
fm_neg, hog_neg_image = hog(neg_img, orientations=9, pixels_per_cell=(8, 8),cells_per_block=(2, 2), visualize=True, multichannel=True)
```

```python
plt.imshow(hog_neg_image)
```

```python
#Visualising ORB feature detection
orb = cv2.ORB_create()
key_points,descriptors = orb.detectAndCompute(img,None)
cv2.drawKeypoints(img,key_points,img)
plt.imshow(img,cmap='gray')
```

```python
#Visualising KAZE feature detection
kaze = cv2.KAZE_create()
key_points,descriptors = kaze.detectAndCompute(img,None)
cv2.drawKeypoints(img,key_points,img)
plt.imshow(img,cmap='gray')
```
